```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = '../data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df.astype(float))
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
#繪製bar chart
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt


plt.bar(df['1stFlrSF'], df['1stFlrSF'], align='center', alpha=0.5)
plt.show()
```

```python
# 將 1stFlrSF 限制在 (minA,maxB)範圍 以內, 調整離群值
minA=500
maxB=2000
df['test1'] = df['1stFlrSF'].clip(minA, maxB)
sns.regplot(x = df['test1'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df.astype(float))
estimator = LinearRegression()
val=cross_val_score(estimator, train_X, train_Y, cv=5).mean()
print("(" + str(minA) + "," + str(maxB) + ") ：" + str(val))

#----------------------#
minA=800
maxB=2500
df['test2'] = df['1stFlrSF'].clip(minA, maxB)
sns.regplot(x = df['test2'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df.astype(float))
estimator = LinearRegression()
val=cross_val_score(estimator, train_X, train_Y, cv=5).mean()
print("(" + str(minA) + "," + str(maxB) + ") ：" + str(val))
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值-->範圍 500~2000
"""
Your Code Here
"""
keep_indexs = (df['1stFlrSF']> 500) & (df['1stFlrSF']< 2000)
df = df[keep_indexs]
train_Y = train_Y[keep_indexs]
sns.regplot(x = df['1stFlrSF'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df.astype(float))
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```
